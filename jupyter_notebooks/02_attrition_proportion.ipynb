```python
#Import necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import os


# Load the raw dataset
df = pd.read_csv("../data/inputs/raw/BankChurners.csv")
df.head()
```

# Proportion of attrited customers and existing customers as a Pie Chart

```python
# Pie chart to show proportion of existing customer and attrited customer
# Example: 'Existing Customer' and 'Attrited Customer'



counts = df.value_counts('Attrition_Flag', normalize=True)



# Assign colors: orange for 'Existing', pink for 'Attrited'
color_map = {
    'Existing Customer': 'orange',
    'Attrited Customer': 'pink'
}
colors = [color_map[label] for label in counts.index]

# Plot the pie chart
plt.figure(figsize=(6, 6))
plt.pie(counts.values, labels=counts.index, colors=colors, autopct='%1.1f%%', startangle=90)
plt.title('Pie Chart of Customer Attrition')
plt.axis('equal')  # Ensures the pie is a circle
plt.show()
```
